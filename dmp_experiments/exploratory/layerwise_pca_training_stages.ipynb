```python
from dmp.data.pmlb import pmlb_loader
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

import sys
sys.path.append("..")

from networks import get_basic_537_houses
from viz import *
```

```python
model, inputs, outputs = get_basic_537_houses(num_hidden_layers=6, hidden_layer_width=128, train_model=False)
```

```python
model.compile(
# loss='binary_crossentropy', # binary classification
# loss='categorical_crossentropy', # categorical classification (one hot)
loss='mean_squared_error',  # regression
optimizer=tf.keras.optimizers.Adam(0.00001),
# optimizer='rmsprop',
# metrics=['accuracy'],
)
```

```python
model.fit(
    x=inputs,
    y=outputs,
    shuffle=True,
    validation_split=.2,
    # epochs=12,
    epochs=5,
    batch_size=256,
    )
```

```python
print_layer_pca(model, inputs, outputs)
```
